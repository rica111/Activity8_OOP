```python
# 3. Applying Inheritance in Python

# Step 1: Define a base class (Parent)
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        return f"Hi, I'm {self.name}."

# Step 2: Define a subclass (Child) that inherits from Person
class Student(Person):
    # This "overrides" the parent's introduce method
    def introduce(self):
        return f"Hi, I'm {self.name}, and I'm a student."

# Step 3: Create objects and call methods
# Using your name for the Student object
person = Person("Allen")
student = Student("Rica Lianne") 

print(person.introduce())
print(student.introduce())
```

```python
# 4. Utilizing Polymorphism in Python

# Step 1: Define a base class
class Shape:
    def area(self):
        return "Calculating area..."

# Step 2: Define subclasses with their own specific implementations
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        # Specific formula for a circle
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        # Specific formula for a rectangle
        return self.length * self.width

# Step 3: Create objects and call their methods
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Both objects use the same method name '.area()', 
# but they produce different results based on their class.
print(f"Circle area: {circle.area()}")
print(f"Rectangle area: {rectangle.area()}")
```

```python
# Try It: Temperature Converter using Static Methods

# Step 1: Define the class
class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        # Using the standard conversion formula
        return (celsius * 9/5) + 32

# Step 2: Prompt the user to enter a temperature in Celsius
# We use float() in case you want to enter decimals
try:
    c_input = float(input("Enter a temperature in Celsius: "))

    # Step 3: Convert using the static method (No object instantiation needed)
    f_result = TemperatureConverter.celsius_to_fahrenheit(c_input)

    # Step 4: Print the result
    print(f"{c_input}°C is equal to {f_result}°F")

except ValueError:
    print("Invalid input! Please enter a numerical value.")
```

```python
# 7. Operator Overloading in Python

# Step 1: Define a class with overloaded operators
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    # Overloading the '+' operator using the __add__ magic method
    def __add__(self, other):
        # This returns a NEW BankAccount object with the combined balance
        return BankAccount(self.balance + other.balance)

# Step 2: Create account objects and add their balances
account1 = BankAccount(5000)
account2 = BankAccount(3000)

# Because we defined __add__, we can use the '+' sign directly on objects
total_balance = account1 + account2  # Combines the balances

# Step 3: Print the result
# Note: Replaced the 'P' in the image with the Peso sign or currency symbol
print(f"Total Balance: ₱{total_balance.balance}")
```

```python
# Part 2: Vector Class with Operator Overloading
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # Overloading the - operator
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    # Helper method to print the vector neatly
    def __str__(self):
        return f"({self.x}, {self.y})"

# Create instances (objects) of the Vector class
v1 = Vector(5, 10)
v2 = Vector(3, 4)

# Demonstrate addition and subtraction using + and -
v_sum = v1 + v2
v_diff = v1 - v2

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Addition (v1 + v2): {v_sum}")
print(f"Subtraction (v1 - v2): {v_diff}")
```

```python
# 8. Abstract Classes and Methods in Python
from abc import ABC, abstractmethod

# Step 1: Define an abstract class
# It inherits from ABC (Abstract Base Class)
class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

# Step 2: Define a subclass implementing the abstract method
class WashingMachine(Appliance):
    def turn_on(self):
        return "Washing machine is now running."

# Step 3: Create an object of the subclass and call the method
wm = WashingMachine()
print(wm.turn_on())
```

```python
# 9. Implementing Method Resolution Order (MRO)

# Step 1: Define multiple classes
class A:
    def show(self):
        return "Class A"

class B(A):
    def show(self):
        return "Class B"

class C(A):
    def show(self):
        return "Class C"

class D(B, C):
    pass # Inherits show() based on MRO

# Step 2: Create an object of class D and call the method
obj = D()
print(obj.show()) # Output depends on MRO

# Bonus: Check the actual MRO list
print(f"MRO for class D: {D.mro()}")
```

```python
# Try It: Method Resolution Order (MRO) with Multiple Inheritance

# Step 1: Define four classes with a show() method
class A:
    def show(self):
        return "Message from Class A"

class B(A):
    def show(self):
        return "Message from Class B"

class C(A):
    def show(self):
        return "Message from Class C"

# Step 2: Define class D that inherits from B and C
class D(B, C):
    def show(self):
        return "Message from Class D"

# Step 3: Explore the MRO
# Printing the MRO list shows the exact path Python takes to find a method
print("Method Resolution Order for Class D:")
print(D.mro())

# Step 4: Create an object of class D and call show()
obj = D()
print(f"\nCalling obj.show(): {obj.show()}")
```

```python
class ATM:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient balance"
        self.balance -= amount
        return f"Withdrawn {amount}, Remaining balance: {self.balance}"

atm = ATM(1000)
print(atm.withdraw(200))
```
